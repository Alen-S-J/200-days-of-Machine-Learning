```python
nlp=spacy.load("en_core_web_sm")
pd.set_option("display.max_rows",200)
```

```python
content = "The growth of Apple Inc is a remarkable tale of innovation and corporate success. Founded in 1976 by Steve Jobs, Steve Wozniak, and Ronald Wayne, Apple started as a player in the personal computer industry. The release of the Macintosh in 1984 marked a turning point, but internal conflicts led to Steve Jobs' departure.The late 1990s saw Apple facing financial challenges until Jobs returned in 1997. The colorful iMac G3 and the Think Different campaign signaled Apple's resurgence. Breakthroughs came with the iPod in 2001, iTunes Store, iPhone in 2007, and the App Store in 2008. These products showcased Apple's design excellence and ability to shape industries.The ecosystem, with seamless device integration and services, contributed to brand loyalty. Financially, Apple became the first trillion-dollar company in 2018. Despite challenges and criticisms, Apple's adaptability, diversified portfolio, and market entry strategies have kept it at the forefront of the tech industry. The story of Apple Inc serves as an inspiration for innovation and corporate success."
 
doc = nlp(content)
 
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

```python
from spacy import displacy
displacy.render(doc, style="ent")
```

```python
entities = [(ent.text, ent.label_, ent.lemma_) for ent in doc.ents]
df = pd.DataFrame(entities, columns=['text', 'type', 'lemma'])
print(df)
```
