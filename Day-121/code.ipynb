### 1. Basics of Time Series Data:


```python
# Import necessary libraries
import pandas as pd

# Creating a basic time series dataset
data = {'Date': pd.date_range('2022-01-01', '2022-01-10'),
        'Value': [10, 12, 14, 18, 22, 25, 28, 32, 36, 40]}

df = pd.DataFrame(data)
print(df)
```

### 2. Trend:


```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating a dataset with a linear trend
data_trend = {'Date': pd.date_range('2022-01-01', '2022-01-10'),
              'Value': [10, 12, 14, 18, 22, 25, 28, 32, 36, 40]}

df_trend = pd.DataFrame(data_trend)

# Plotting the trend
plt.plot(df_trend['Date'], df_trend['Value'])
plt.title('Time Series Data with Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```

### 3. Seasonality:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Creating a dataset with seasonality
data_seasonal = {'Date': pd.date_range('2022-01-01', '2022-01-30', freq='D'),
                 'Value': 20 + 10 * np.sin(np.arange(30) * (2 * np.pi / 7))}

df_seasonal = pd.DataFrame(data_seasonal)

# Plotting the seasonality
plt.plot(df_seasonal['Date'], df_seasonal['Value'])
plt.title('Time Series Data with Seasonality')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```

### 4. Noise:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Creating a dataset with noise
np.random.seed(42)
data_noise = {'Date': pd.date_range('2022-01-01', '2022-01-30', freq='D'),
              'Value': 20 + 10 * np.sin(np.arange(30) * (2 * np.pi / 7)) + np.random.normal(0, 5, 30)}

df_noise = pd.DataFrame(data_noise)

# Plotting the data with noise
plt.plot(df_noise['Date'], df_noise['Value'])
plt.title('Time Series Data with Noise')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```
